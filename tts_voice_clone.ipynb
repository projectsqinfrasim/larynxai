```python
import requests

API_KEY = "x"

url = "https://api.gradium.ai/api/post/speech/tts"

headers = {
    "x-api-key": API_KEY,
    "Content-Type": "application/json"
}

payload = {
    "text": "Hello, this is my first Gradium API test.",
    "voice_id": "Zd5POlBGSbD-JBXF",
    "output_format": "wav",
    "only_audio": True
}


response = requests.post(url, headers=headers, json=payload)

print("status:", response.status_code)
print("content-type:", response.headers.get("content-type"))

if response.status_code == 200:
    with open("output.wav", "wb") as f:
        f.write(response.content)
    print("audio saved as output.wav")
else:
    print(response.text)
```

```python
url = "https://api.gradium.ai/api/voices/"

headers = {
    "x-api-key": API_KEY
}

files = {
    "audio_file": open("output.wav", "rb")
}

data = {
    "name": "my_voice",
    "input_format": "wav",
    "description": "test voice for hackathon",
    "language": "en",
    "start_s": 0,
    "timeout_s": 10
}

response = requests.post(
    url,
    headers=headers,
    files=files,
    data=data
)

print("status:", response.status_code)
print(response.text)
```
